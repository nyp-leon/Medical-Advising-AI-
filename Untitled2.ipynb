```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
df = pd.read_excel('Desktop/FYPJ Dataset.xlsx', sheet_name = 'Sheet1')
df = pd.DataFrame({'Id' : df['Unique_Identifier'], 'Date' : df['Test_Date'], 'Score' : df['Score']})
df.head()
```

```python
testn = pd.read_excel('Desktop/testoutput3.xlsx', sheet_name = 'Sheet3')
testn = pd.DataFrame({ 'Id' : testn['Id'],  
                         'Date' : testn['Date'], 
                         'Score' : testn['Glucose']})
testn.head()
```

```python
testn = testn.sort_values(['Id', 'Date'], ascending=['True', 'True'])
testn.head()
```

```python
testn.Date = pd.to_datetime(testn.Date)
#testn.set_index(['Date'], inplace=True)

testn.set_index(['Id'], inplace=True)
testn
```

```python
plt.figure(figsize=(30, 15))
plt.plot(testn['Date'], testn['Score'], label = 'a')
plt.xlabel('Date'); plt.ylabel('a'); plt.title('a')
plt.legend();
```

```python
ids = testn["Id"]
testn = testn[ids.isin(ids[ids.duplicated()])].sort_values("Id")
```

```python
testn = testn.pivot_table(index='Date', columns='Id', values='Score')
testn
```
