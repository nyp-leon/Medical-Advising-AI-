```python
y = x_train1['Score'].resample('MS').mean()
```

```python
y['2016']
```

```python
#Fit dataset and ideal parameter combination into SARIMAX model
#Display results
mod = ARIMA(y['2016':'2017'],
            order=(5, 1, 0)
            )
results = mod.fit()
print(results.summary())
```

```python
residuals = DataFrame(results.resid)
residuals.plot()
plt.title('Residual Error Line Plot');
plt.show()

residuals.plot(kind='kde')
plt.title('Residual Error Density Plot');
plt.show()
print(residuals.describe())
```

```python
X = x_train1.groupby(['ID', 'Test_Date','Name']).mean()['Score'].values

size = int(len(X) * 0.66)
train, test = X[1:len(X)-7], X[len(X)-7:]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```
