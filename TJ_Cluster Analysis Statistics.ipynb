```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
#Import Excel And specify which sheet name to use for which dataframe as there are 2 sheets with different clustering results
data = pd.read_excel('C:/Users/L30908/Cluster Results.xlsx', header=0, sheet_name='KMeans_Cluster')
data2 = pd.read_excel('C:/Users/L30908/Cluster Results.xlsx', header=0, sheet_name='GMM_Cluster')

df1 = pd.DataFrame({'ID': data['ID'], 'Test_Date': data['Test_Date'], 'Age': data['Age'], 'Name': data['Name'], 'Score': data['Score'], 'kmeans': data['kmeans'], 'Score_Level': data['Score_Level'], 'T1': data['T1'],'T2': data['T2'],'SI_Conversion': data['SI_Conversion']})
df2 = pd.DataFrame({'ID': data2['ID'], 'Test_Date': data2['Test_Date'], 'Age': data2['Age'], 'Name': data2['Name'], 'Score': data2['Score'], 'gmm': data2['gmm'], 'Score_Level': data2['Score_Level'], 'T1': data2['T1'],'T2': data2['T2'],'SI_Conversion': data2['SI_Conversion']})
```

```python
#View number of records for each kmeans cluster
ax = df1.groupby('kmeans').count()['Score'].plot(figsize=(20,10), kind='bar', fontsize=20)
plt.xlabel('kmeans Cluster', fontsize=20);
plt.ylabel('Count', fontsize=20);
# create a list to collect the plt.patches data
totals = []

# find the values and append to list
for i in ax.patches:
    totals.append(i.get_height())

# set individual bar lables using above list
total = sum(totals)

# set individual bar lables using above list
for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
    ax.text(i.get_x()+.1, i.get_height(), \
            str(round(i.get_height(), 2)), fontsize=22,
                color='black')
```

```python
#Calculate mean value for each kmeans cluster
ax = df1.groupby('kmeans').mean()['Score'].plot(figsize=(20,10), kind='bar', fontsize=20)
plt.xlabel('kmeans Cluster', fontsize=20);
plt.ylabel('Average Score', fontsize=20);
# create a list to collect the plt.patches data
totals = []

# find the values and append to list
for i in ax.patches:
    totals.append(i.get_height())

# set individual bar lables using above list
total = sum(totals)

# set individual bar lables using above list
for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
    ax.text(i.get_x()+.1, i.get_height(), \
            str(round(i.get_height(), 2)), fontsize=22,
                color='black')
```
